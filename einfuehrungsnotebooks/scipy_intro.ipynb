# Was ist SciPy?

SciPy (Scientific Python) ist eine umfangreiche Sammlung von Algorithmen und mathematischen Werkzeugen, die auf NumPy aufbaut. Sie erweitert die Möglichkeiten von Python im Bereich des wissenschaftlichen Rechnens erheblich. <br>

Wichtige Merkmale von SciPy sind: <br>
- Umfassende Sammlung wissenschaftlicher Algorithmen <br>
- Optimierte und getestete Implementierungen mathematischer Routinen <br>
- Modulare Struktur für verschiedene Anwendungsbereiche <br>
- Nahtlose Integration mit NumPy <br>
- Aktive wissenschaftliche Community <br>

Die wichtigsten Module von SciPy sind: <br>
- **scipy.optimize**: Algorithmen für Optimierung und Wurzelsuche <br>
- **scipy.integrate**: Routinen für numerische Integration <br>
- **scipy.linalg**: Erweiterte lineare Algebra-Funktionen <br>
- **scipy.signal**: Signalverarbeitung und Filterdesign <br>
- **scipy.stats**: Statistische Verteilungen und Funktionen <br>
- **scipy.ndimage**: Mehrdimensionale Bilddatenverarbeitung <br>

SciPy ist besonders wichtig in: <br>
- Wissenschaftlicher Forschung <br>
- Ingenieurwesen <br>
- Datenanalyse <br>
- Signalverarbeitung <br>
- Bildverarbeitung <br>

Die Bibliothek ist zum Standard in der wissenschaftlichen Python-Community geworden, da sie zuverlässige, effiziente und gut dokumentierte Implementierungen wichtiger Algorithmen bietet. <br>

## NumPy vs. SciPy: Was ist der Unterschied?

NumPy und SciPy sind eng miteinander verbunden, haben aber unterschiedliche Schwerpunkte: <br>

### NumPy bietet:
- Grundlegende Array-Operationen <br>
- Elementweise mathematische Operationen <br>
- Einfache lineare Algebra <br>
- Basis-Funktionen für zufällige Zahlen <br>
- Grundlegende statistische Operationen <br>

### SciPy baut auf NumPy auf und bietet:
- Fortgeschrittene wissenschaftliche Algorithmen <br>
- Komplexe mathematische Routinen <br>
- Spezialisierte Anwendungen <br>

### Wann welche Bibliothek?

**NumPy, wenn:**
- Du hauptsächlich mit Arrays arbeitest <br>
- Du grundlegende mathematische Operationen brauchst <br>
- Du elementweise Berechnungen durchführst <br>
- Du einfache lineare Algebra benötigst <br>

**SciPy, wenn:**
- Du spezielle wissenschaftliche Algorithmen brauchst <br>
- Du numerische Integration durchführst <br>
- Du Optimierungsprobleme löst <br>
- Du fortgeschrittene statistische Tests benötigst <br>
- Du Signalverarbeitung implementierst <br>
- Du sparse Matrizen verarbeitest <br>

### Zusammenspiel:

In der Praxis werden beide Bibliotheken oft zusammen verwendet: <br>
- NumPy für die grundlegende Datenstruktur (Arrays) <br>
- SciPy für fortgeschrittene Berechnungen auf diesen Arrays <br>



## 1 Optimierung

SciPy bietet verschiedene Funktionen zur Optimierung: <br>

```python
from scipy import optimize
import numpy as np

# Funktion definieren die minimiert werden soll
def f(x):
    return (x[0] - 2)**2 + (x[1] - 3)**2

# Startpunkt
x0 = [0, 0]

# Minimierung
ergebnis = optimize.minimize(f, x0)
print("Minimierungsergebnis:")
print(f"x = {ergebnis.x[0]:.2f}, y = {ergebnis.x[1]:.2f}")
print(f"Minimaler Wert = {ergebnis.fun:.2f}")

# Wurzelsuche
def g(x):
    return x**2 - 4

wurzel = optimize.root(g, x0=3)
print("\nWurzel von x²-4:", wurzel.x[0])
```

#### Erklärung:

**`scipy.optimize.minimize()`**

Diese Funktion findet das Minimum einer skalaren Funktion mit einer oder mehreren Variablen.

**Parameter:**
- `fun`: Die zu minimierende Zielfunktion
- `x0`: Startpunkt für die Optimierung (Array oder Skalar)
- `method` (optional): Optimierungsalgorithmus (Standard: automatische Auswahl)

**Rückgabe:** OptimizeResult-Objekt mit:
- `x`: Koordinaten des gefundenen Minimums
- `fun`: Funktionswert am Minimum
- `success`: Boolean, ob Optimierung erfolgreich war

**`scipy.optimize.root()`**

Diese Funktion findet die Nullstellen (Wurzeln) einer skalaren oder vektoriellen Funktion.

**Parameter:**
- `fun`: Die Funktion, deren Nullstellen gesucht werden
- `x0`: Startpunkt für die Wurzelsuche
- `method` (optional): Algorithmus für die Wurzelsuche (Standard: 'hybr')

**Rückgabe:** OptimizeResult-Objekt mit:
- `x`: Koordinaten der gefundenen Nullstelle
- `fun`: Funktionswert an der Nullstelle (sollte ≈ 0 sein)
- `success`: Boolean, ob die Suche erfolgreich war

## 2 Integration und Differentiation

Numerische Integration und Differentiation: <br>

```python
from scipy import integrate

# Integration
def h(x):
    return np.sin(x)**2

# Bestimmtes Integral
ergebnis, fehler = integrate.quad(h, 0, np.pi)
print(f"Integral von sin²(x) von 0 bis π = {ergebnis:.6f}")

# Mehrfachintegration
def f2(x, y):
    return x * y

ergebnis, fehler = integrate.dblquad(f2, 0, 1,    # y-Grenzen
                                    lambda x: 0,    # x-Untergrenze
                                    lambda x: 1)    # x-Obergrenze
print(f"\nDoppelintegral = {ergebnis:.6f}")
```

#### Erklärung:

**`scipy.integrate.quad()`**

Diese Funktion berechnet bestimmte Integrale einer skalaren Funktion mit einer Variable mittels adaptiver Quadratur.

**Parameter:**
- `func`: Die zu integrierende Funktion
- `a`: Untere Integrationsgrenze
- `b`: Obere Integrationsgrenze
- `args` (optional): Zusätzliche Parameter für die Funktion

**Rückgabe:** Tupel aus:
- `integral`: Numerischer Wert des berechneten Integrals
- `abserr`: Geschätzter absoluter Fehler des Ergebnisses

**`scipy.integrate.dblquad()`**

Diese Funktion berechnet Doppelintegrale (zweifache Integration) über rechteckige oder beliebige Bereiche.

**Parameter:**
- `func`: Die zu integrierende Funktion f(y, x) - Achtung: y ist der erste Parameter!
- `a`, `b`: Grenzen für die äußere Integration (meist x-Richtung)
- `gfun`, `hfun`: Funktionen oder Konstanten für die inneren Integrationsgrenzen (y-Richtung)

**Rückgabe:** Tupel aus:
- `integral`: Numerischer Wert des berechneten Doppelintegrals
- `abserr`: Geschätzter absoluter Fehler des Ergebnisses

## 3 Lineare Algebra

Erweiterte lineare Algebra-Funktionen: <br>

```python
from scipy import linalg

# Matrix erstellen
A = np.array([[1, 2], [3, 4]])
b = np.array([5, 6])

# Lineares Gleichungssystem lösen
x = linalg.solve(A, b)
print("Lösung des Gleichungssystems:")
print(x)

# Eigenwerte und Eigenvektoren
eigvals, eigvecs = linalg.eig(A)
print("\nEigenwerte:")
print(eigvals)

# LU-Zerlegung
P, L, U = linalg.lu(A)
print("\nLU-Zerlegung - U-Matrix:")
print(U)
```

#### Erklärung:

**`scipy.linalg.solve()`**

Diese Funktion löst lineare Gleichungssysteme der Form Ax = b, wobei A eine quadratische Matrix ist.

**Parameter:**
- `a`: Koeffizientenmatrix A (quadratisch, n×n)
- `b`: Rechte Seite des Gleichungssystems (Vektor der Länge n)
- `assume_a` (optional): Annahmen über Matrix A (z.B. 'pos' für positiv definit)

**Rückgabe:**
- `x`: Lösungsvektor des Gleichungssystems

**`scipy.linalg.eig()`**

Diese Funktion berechnet Eigenwerte und Eigenvektoren einer quadratischen Matrix.

**Parameter:**
- `a`: Quadratische Matrix (n×n)
- `b` (optional): Zweite Matrix für verallgemeinerte Eigenprobleme
- `left`, `right` (optional): Bestimmt, welche Eigenvektoren berechnet werden

**Rückgabe:** Tupel aus:
- `w`: Array der Eigenwerte
- `v`: Matrix der Eigenvektoren (Spalten sind die Eigenvektoren)

**`scipy.linalg.lu()`**

Diese Funktion führt eine LU-Zerlegung durch: PA = LU (mit Pivotisierung).

**Parameter:**
- `a`: Matrix, die zerlegt werden soll
- `permute_l` (optional): Bestimmt Form der Ausgabe

**Rückgabe:** Tupel aus:
- `P`: Permutationsmatrix
- `L`: Untere Dreiecksmatrix (Lower triangular)
- `U`: Obere Dreiecksmatrix (Upper triangular)

## 4 Signalverarbeitung

Das Signal-Modul für Signalverarbeitung: <br>

```python
from scipy import signal
import numpy as np

# Signal generieren
t = np.linspace(0, 1, 1000)
x = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*10*t)

# Filter entwerfen
b, a = signal.butter(4, 0.2, 'low')
y = signal.filtfilt(b, a, x)

# Spektralanalyse
f, Pxx = signal.welch(x, fs=1000)

# Fensterfunktionen
fenster = signal.windows.hann(50)
print("Hann-Fenster (erste 5 Werte):")
print(fenster[:5])
```

#### Erklärung:

**`scipy.signal.butter()`**

Diese Funktion entwirft einen digitalen Butterworth-Filter (IIR-Filter mit maximal flachem Frequenzgang).

**Parameter:**
- `N`: Filterordnung (bestimmt Steilheit des Übergangs)
- `Wn`: Kritische Frequenz(en) normalisiert auf Nyquist-Frequenz (0-1)
- `btype` (optional): Filtertyp ('low', 'high', 'band', 'bandstop')
- `analog` (optional): Wenn False (Standard), digitaler Filter

**Rückgabe:** Tupel aus:
- `b`: Zählerkoeffizienten der Übertragungsfunktion
- `a`: Nennerkoeffizienten der Übertragungsfunktion

**`scipy.signal.filtfilt()`**

Diese Funktion wendet einen digitalen Filter vorwärts und rückwärts an (Zero-Phase-Filterung).

**Parameter:**
- `b`: Zählerkoeffizienten des Filters
- `a`: Nennerkoeffizienten des Filters  
- `x`: Eingangssignal (Array)
- `axis` (optional): Achse entlang der gefiltert wird

**Rückgabe:**
- `y`: Gefiltertes Signal ohne Phasenverschiebung

**`scipy.signal.welch()`**

Diese Funktion schätzt die Leistungsspektraldichte mittels Welch-Methode (überlappende Segmente).

**Parameter:**
- `x`: Eingangssignal
- `fs` (optional): Abtastfrequenz (Standard: 1.0)
- `nperseg` (optional): Länge der Segmente
- `noverlap` (optional): Überlappung zwischen Segmenten

**Rückgabe:** Tupel aus:
- `f`: Array der Frequenzwerte
- `Pxx`: Leistungsspektraldichte

**`scipy.signal.windows.hann()`**

Diese Funktion erzeugt ein Hann-Fenster (auch Hanning-Fenster genannt).

**Parameter:**
- `M`: Anzahl der Punkte im Fenster
- `sym` (optional): Wenn True (Standard), symmetrisches Fenster

**Rückgabe:**
- `w`: Hann-Fenster als Array

## 5 Statistik

Statistische Funktionen und Tests: <br>

```python
from scipy import stats

# Normalverteilte Daten generieren
daten = np.random.normal(loc=0, scale=1, size=1000)

# Deskriptive Statistik
print("Normalverteilungstest:")
print(stats.normaltest(daten))

# t-Test
gruppe1 = np.random.normal(0, 1, 100)
gruppe2 = np.random.normal(0.5, 1, 100)
t_stat, p_wert = stats.ttest_ind(gruppe1, gruppe2)
print("\nt-Test p-Wert:", p_wert)

# Korrelation
x = np.random.normal(0, 1, 100)
y = 2*x + np.random.normal(0, 0.5, 100)
r, p = stats.pearsonr(x, y)
print("\nPearson Korrelation:", r)
```

#### Erklärung:

**`scipy.stats.normaltest()`**

Diese Funktion testet, ob eine Stichprobe aus einer Normalverteilung stammt (D'Agostino und Pearson's Test).

**Parameter:**
- `a`: Eingangsdaten (Array)
- `axis` (optional): Achse entlang der getestet wird
- `nan_policy` (optional): Umgang mit NaN-Werten ('propagate', 'raise', 'omit')

**Rückgabe:** Tupel aus:
- `statistic`: Teststatistik
- `pvalue`: p-Wert (wenn > 0.05, dann normalverteilt bei α=0.05)

**`scipy.stats.ttest_ind()`**

Diese Funktion führt einen unabhängigen Zweistichproben-t-Test durch (prüft Gleichheit der Mittelwerte).

**Parameter:**
- `a`: Erste Stichprobe (Array)
- `b`: Zweite Stichprobe (Array)
- `equal_var` (optional): Annahme gleicher Varianzen (Standard: True)
- `alternative` (optional): Art des Tests ('two-sided', 'less', 'greater')

**Rückgabe:** Tupel aus:
- `statistic`: t-Statistik
- `pvalue`: p-Wert des Tests

**`scipy.stats.pearsonr()`**

Diese Funktion berechnet den Pearson-Korrelationskoeffizienten und testet auf lineare Abhängigkeit.

**Parameter:**
- `x`: Erste Variable (Array)
- `y`: Zweite Variable (Array)
- `alternative` (optional): Art des Tests für Signifikanz

**Rückgabe:** Tupel aus:
- `statistic`: Pearson-Korrelationskoeffizient r (-1 bis +1)
- `pvalue`: p-Wert für Test auf Signifikanz der Korrelation

# Mehrdimensionale Bilddaten

Funktionen zum verarbeiten und modifizieren von Bilddaten

```python
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt

# Beispielbild erstellen (2D Array)
bild = np.zeros((50, 50))
bild[15:35, 15:35] = 1  # Quadrat in der Mitte
bild[10:15, 40:45] = 1  # Kleines zusätzliches Quadrat

# Plot-Setup
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
fig.suptitle('SciPy NDImage - Bildverarbeitungsschritte', fontsize=16)

# 1. Originalbild
axes[0,0].imshow(bild, cmap='gray')
axes[0,0].set_title('Original')
axes[0,0].axis('off')

# 2. Gaussian Filter
bild_geglättet = ndimage.gaussian_filter(bild, sigma=2)
axes[0,1].imshow(bild_geglättet, cmap='gray')
axes[0,1].set_title('Gaussian Filter (σ=2)')
axes[0,1].axis('off')
print("Gaussian Filter angewendet")

# 3. Morphologische Operationen
struktur = np.ones((3,3))  # 3x3 Strukturelement
bild_opening = ndimage.binary_opening(bild, structure=struktur)
bild_closing = ndimage.binary_closing(bild, structure=struktur)

axes[0,2].imshow(bild_opening, cmap='gray')
axes[0,2].set_title('Binary Opening')
axes[0,2].axis('off')

axes[0,3].imshow(bild_closing, cmap='gray')
axes[0,3].set_title('Binary Closing')
axes[0,3].axis('off')

# 4. Objekterkennung und Messung
labeled_bild, anzahl_objekte = ndimage.label(bild)
axes[1,0].imshow(labeled_bild, cmap='nipy_spectral')
axes[1,0].set_title(f'Labeled Objects ({anzahl_objekte})')
axes[1,0].axis('off')
print(f"Anzahl gefundener Objekte: {anzahl_objekte}")

# 5. Schwerpunkt berechnen und visualisieren
schwerpunkt = ndimage.center_of_mass(bild)
axes[1,1].imshow(bild, cmap='gray')
axes[1,1].plot(schwerpunkt[1], schwerpunkt[0], 'r+', markersize=15, markeredgewidth=3)
axes[1,1].set_title(f'Schwerpunkt\n({schwerpunkt[0]:.1f}, {schwerpunkt[1]:.1f})')
axes[1,1].axis('off')
print(f"Schwerpunkt: ({schwerpunkt[0]:.1f}, {schwerpunkt[1]:.1f})")

# 6. Geometrische Transformationen
bild_rotiert = ndimage.rotate(bild, angle=45, reshape=False)
axes[1,2].imshow(bild_rotiert, cmap='gray')
axes[1,2].set_title('Rotiert (45°)')
axes[1,2].axis('off')

bild_vergrößert = ndimage.zoom(bild, zoom=1.5)
axes[1,3].imshow(bild_vergrößert, cmap='gray')
axes[1,3].set_title('Vergrößert (1.5x)')
axes[1,3].axis('off')

# 7. Distanztransformation
distanz = ndimage.distance_transform_edt(bild)
im_distanz = axes[2,0].imshow(distanz, cmap='hot')
axes[2,0].set_title(f'Distanztransformation\n(Max: {np.max(distanz):.2f})')
axes[2,0].axis('off')
plt.colorbar(im_distanz, ax=axes[2,0], shrink=0.8)
print(f"Maximale Distanz: {np.max(distanz):.2f}")

# 8. Erosion und Dilation zum Vergleich
bild_erosion = ndimage.binary_erosion(bild, structure=struktur)
bild_dilation = ndimage.binary_dilation(bild, structure=struktur)

axes[2,1].imshow(bild_erosion, cmap='gray')
axes[2,1].set_title('Binary Erosion')
axes[2,1].axis('off')

axes[2,2].imshow(bild_dilation, cmap='gray')
axes[2,2].set_title('Binary Dilation')
axes[2,2].axis('off')

# 9. Kantendetektion mit Sobel-Filter
kanten = ndimage.sobel(bild_geglättet)
axes[2,3].imshow(kanten, cmap='gray')
axes[2,3].set_title('Sobel Kantendetektion')
axes[2,3].axis('off')

plt.tight_layout()
plt.show()

# Zusätzliche Informationen ausgeben
print(f"\nBildgrößen:")
print(f"Original: {bild.shape}")
print(f"Vergrößert: {bild_vergrößert.shape}")
print(f"Rotiert: {bild_rotiert.shape}")
```

#### Erklärung:

**`scipy.ndimage.gaussian_filter()`**

Diese Funktion wendet einen Gaussian-Filter (Weichzeichnung) auf mehrdimensionale Arrays an.

**Parameter:**
- `input`: Eingangsbild/-array
- `sigma`: Standardabweichung des Gaussian-Kernels (bestimmt Stärke der Glättung)
- `order` (optional): Ordnung der Ableitung (0 = normale Filterung)
- `mode` (optional): Behandlung der Bildränder ('reflect', 'constant', etc.)

**Rückgabe:**
- Gefiltertes Array mit gleicher Form wie Eingabe

**`scipy.ndimage.binary_opening()` / `scipy.ndimage.binary_closing()`**

Diese Funktionen führen morphologische Operationen auf binäre Bilder durch.

**Parameter:**
- `input`: Binäres Eingangsbild (0 und 1 Werte)
- `structure` (optional): Strukturelement für die Operation
- `iterations` (optional): Anzahl der Wiederholungen

**Rückgabe:**
- Binäres Array nach morphologischer Operation
- **Opening**: Entfernt kleine Objekte und glättet Konturen
- **Closing**: Schließt kleine Löcher in Objekten

**`scipy.ndimage.label()`**

Diese Funktion identifiziert und nummeriert zusammenhängende Regionen in binären Bildern.

**Parameter:**
- `input`: Binäres Eingangsbild
- `structure` (optional): Definition der Nachbarschaft (Standard: vollständige Konnektivität)

**Rückgabe:** Tupel aus:
- `labeled_array`: Array mit nummerierten Objekten (0=Hintergrund, 1,2,3...=Objekte)
- `num_features`: Anzahl der gefundenen Objekte

**`scipy.ndimage.center_of_mass()`**

Diese Funktion berechnet den Schwerpunkt (Zentrum der Masse) von Objekten.

**Parameter:**
- `input`: Eingangsbild/-array
- `labels` (optional): Gelabeltes Bild für spezifische Objekte
- `index` (optional): Welche Labels berücksichtigt werden sollen

**Rückgabe:**
- Koordinaten des Schwerpunkts als Tupel

**`scipy.ndimage.rotate()` / `scipy.ndimage.zoom()`**

Diese Funktionen führen geometrische Transformationen durch.

**Rotate Parameter:**
- `input`: Eingangsbild
- `angle`: Rotationswinkel in Grad
- `reshape` (optional): Ob Ausgabeform angepasst wird

**Zoom Parameter:**
- `input`: Eingangsbild
- `zoom`: Skalierungsfaktor (>1 vergrößert, <1 verkleinert)

**`scipy.ndimage.distance_transform_edt()`**

Diese Funktion berechnet die euklidische Distanztransformation.

**Parameter:**
- `input`: Binäres Bild (True/1 = Objekt, False/0 = Hintergrund)
- `sampling` (optional): Pixelabstände in jeder Dimension

**Rückgabe:**
- Array mit Distanzwerten (Abstand zum nächsten Hintergrundpixel)
